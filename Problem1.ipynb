#### Functions

```python
def do_dist(series_data, col_name=''):
    """
    For Continious variables
    """
    plt.figure(figsize=(15,10))
    plt.subplot(2,1,1)
    sns.distplot(series_data,bins=50,color='blue')
    plt.title(f'Distribution of column {col_name}',{'color' :'black','fontsize':14})
    plt.xlabel(f'{col_name}')
    plt.show()

def do_count(series_data, col_name=''):
    """
    For Counting variables
    """
    plt.figure(figsize=(15,10))
    plt.subplot(2,1,1)
    sns.countplot(series_data)
    plt.title(f'Distribution of column {col_name}',{'color' :'black','fontsize':14})
    plt.xlabel(f'{col_name}')
    plt.show()
    
# Label Encoding
def do_label_encoding(train_data, test_data, cat_cols):
    for col in cat_cols:
        le = LabelEncoder()
        train_data[col] = le.fit_transform(train_data[col])
        test_data[col] =  le.transform(test_data[col])
    return train_data, test_data  

def eval_model(y_true, y_pred):
    conf_mat = metrics.confusion_matrix(y_test, y_pred)
    acc_score = metrics.accuracy_score(y_test, y_pred)
    prec = metrics.precision_score(y_test, y_pred)
    rec = metrics.recall_score(y_test, y_pred)
    print(f"Accuracy Score: {acc_score}")
    print(f"Precision: {prec}")
    print(f"Recall: {rec}")
    print("Confusion Matrix")
    sns.heatmap(conf_mat, annot=True, cmap='Blues', fmt='g')
    plt.show()    
```

#### Feature Engineering

```python
# Observed space in strings - Remove
data = data.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
# Replace the output col cat into 0 and 1
# 0: <=50K
# 1: >50K
op_replace_di = {'<=50K': 0, '>50K': 1}
data['salary'] = data['salary'].replace(op_replace_di)
```

```python
do_count(data['salary'], 'salary')
```

```python
# Observed ? in few cols. Replace by null and fill by median for safe side
data.isin(['?']).sum()
```

```python
data = data.replace('?', np.NaN)
data.info()
```

```python
def smart_fill_na(data):
    obj_cols = {k:v for k, v in data.dtypes.iteritems() if v == 'O'}
    for col in obj_cols.keys():
        fill_val = data[col].mode()[0]
        print(f"Col: {col}, Fill Value: {fill_val}")
        data[col] = data[col].fillna(fill_val)
    return data
```

```python
data = smart_fill_na(data)
data.info()
```

```python
# Reduce Cat
data['marital-status'].unique()
```

```python
data['marital-status'] = data['marital-status'].replace(['Widowed', 'Divorced', 'Separated', 'Never-married'], 'single')
data['marital-status'] = data['marital-status'].replace(['Married-spouse-absent', 'Married-civ-spouse', 'Married-AF-spouse'], 'married')
```

```python
do_count(data['marital-status'], 'marital-status')
```

```python
# Put last .1% countries to Other
threshold_percent = 0.1
vc = pd.value_counts(data['country'])
mask = (vc / vc.sum() * 100).lt(threshold_percent)
data = data.assign(country = np.where(data['country'].isin(vc[mask].index),'Other', data['country']))
```

```python
do_count(data['country'], 'country')
```

```python
# workclass
map_di = {'Without-pay': 'nopay', 'Never-worked': 'nopay'}
data['workclass'] = data['workclass'].replace(map_di)
do_count(data['workclass'], 'workclass')
```

```python
# Make a copy of prepared data
prep_data = data.copy(deep=True)
```

```python
prep_data.info()
```

```python
X = prep_data.drop('salary', axis=1)
y = prep_data['salary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, stratify=y, random_state=100)
```

```python
cat_cols = {k:v for k, v in data.dtypes.iteritems() if v == 'O'}
```

```python
X_train_le, X_test_le = do_label_encoding(X_train, X_test, cat_cols=cat_cols)
```

```python
eval_model(y_test, y_pred)
```

```python
def do_parameters_search(model, param_dict, x_train, y_train, nt=100):
    print(f"Model passed: {model}")
    cv = StratifiedKFold(n_splits=5, random_state=1, shuffle=True)
    random_search_obj = RandomizedSearchCV(estimator = model, param_distributions = param_dict, n_iter = nt, \
                                           cv = cv,verbose = 1, random_state = 10,scoring = 'accuracy', 
                                           return_train_score = True, n_jobs = -1)
    random_search_obj.fit(x_train,y_train)
    print(f"Mean_train_score is: {np.mean(random_search_obj.cv_results_['mean_train_score']):.3f}")
    print(f"Mean_test_score is: {np.mean(random_search_obj.cv_results_['mean_test_score']):.3f}")
    print(f"Best parameters are: {random_search_obj.best_params_}")
    return random_search_obj.best_params_
```

```python
rf_param_dict = {
    'n_estimators': np.arange(100, 400, 20),
    'min_samples_leaf': np.arange(20, 50, 5)
}
do_parameters_search(rfc, rf_param_dict, X_train_le, y_train)
```
